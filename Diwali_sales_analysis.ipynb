```python
Diwali_sales.groupby(['Marital_Status','Gender'])['Amount'].sum() # gives count of amount with respect to marital status and gender.
```

```python
Diwali_sales.groupby(['Marital_Status','Gender'])['Amount'].sum().reset_index()   # reset function converts the series into data frame
```

```python
sales_marital_status = Diwali_sales.groupby(['Marital_Status','Gender'])['Amount'].sum().reset_index()
```

```python
sns.barplot(x = 'Marital_Status', y = 'Amount',hue = 'Gender', data = sales_marital_status)
plt.title('total amt spent by marital status w.r.t Gender')
plt.xlabel('Marital Status')
plt.ylabel('Total amount spent')
plt.show
```

the data suggests that female customers represent a critical target demographic for revenue growth due to their consistently higher spending patterns.
